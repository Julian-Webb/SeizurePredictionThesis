Compare all seizures to see if any are duplicated between patients

```python
from config.paths import PATHS, Dataset
from utils.io import pickle_path

anns = {}
for pt in PATHS.patient_dirs([Dataset.for_mayo], True):
    anns[pt.name] = pickle_path(pt.szr_anns_original_dir / pt.name)
for pt in PATHS.patient_dirs([Dataset.uneeg_extended], True):
    anns[pt.name] = pickle_path(pt.combined_anns_file)
for pt in PATHS.patient_dirs([Dataset.competition], True):
    anns[pt.name] = pickle_path(pt.all_szr_starts_file)
anns
```

```python
import pandas as pd
from pandas import DataFrame

dfs = []
keys = []

for ptnt, ann in anns.items():
    df = pd.read_pickle(ann)
    dfs.append(df)
    keys.append(ptnt)

all_szrs = pd.concat(dfs, keys=keys, names=['patient', 'seizure_idx'])
all_szrs = all_szrs.drop(columns=['valid', 'Unnamed: 0'])
all_szrs
```

```python
duplicates = {}

for col in ['single_marker', 'start', 'end']:
    duplicate_mask = all_szrs[col].duplicated(keep=False)
    dups = all_szrs[col][duplicate_mask]
    dups.dropna(inplace=True)
    duplicates[col] = dups
```
