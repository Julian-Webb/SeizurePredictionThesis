```python


import pandas as pd
from pandas import DataFrame

from config.paths import PATHS
from utils.io import pickle_path
```

```python
# ptnt_dirs = [PatientDir(Path('/Users/julian/Developer/SeizurePredictionData/20240201_UNEEG_ForMayo/ptnt1'))]
ptnt_dirs = PATHS.patient_dirs()
```

```python
stats = {}

for ptnt_dir in ptnt_dirs:
    segs = pd.read_pickle(pickle_path(ptnt_dir.segments_table))
    counts = segs['type'].value_counts()
    ratio = counts['preictal'] / counts['interictal']
    stats[ptnt_dir.name] = {**counts, 'preictal to interictal ratio': ratio}
```
