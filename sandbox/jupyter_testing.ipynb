```python
# python
import sys, os
from pathlib import Path

print("cwd:", Path.cwd())
print("\nPYTHONPATH env:", os.environ.get("PYTHONPATH"))

print("\nsys.path entries:")
for i, p in enumerate(sys.path):
    print(f"{i}: {p}")

print("\nentries containing `config/paths.py`:")
found = False
for i, p in enumerate(sys.path):
    try:
        if Path(p).joinpath("config", "paths.py").exists():
            print(f"{i}: {p}")
            found = True
    except Exception:
        pass
if not found:
    print("none found")
```
