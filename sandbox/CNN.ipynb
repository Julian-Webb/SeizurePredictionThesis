```python
# 2D Model: CNN with independent channels and batch norm
n_samples, n_channels = 3105, 2

model = tf.keras.models.Sequential([
    # Arguments:
    # Conv1D(filters, kernel_size, ...)
    # MaxPool1D(pool_size, strides, padding, ...)
    Input([n_samples, n_channels, 1]),
    BatchNormalization(name='batch_norm0'),

    Conv2D(32, [5, 1], padding='same', activation='relu', name='conv1'),
    BatchNormalization(name='batch_norm1'),
    MaxPool2D([5, 1], padding='same', name='pool1'),
    Dropout(0.2, name='dropout1'),

    Conv2D(64, [5, 1], padding='same', activation='relu', name='conv2'),
    BatchNormalization(name='batch_norm2'),
    MaxPool2D([3, 1], padding='same', name='pool2'),
    Dropout(0.2, name='dropout2'),

    Conv2D(96, [3, 1], padding='same', activation='relu', name='conv3'),
    BatchNormalization(name='batch_norm3'),
    MaxPool2D([2, 1], padding='same', name='pool3'),
    Dropout(0.2, name='dropout3'),

    Conv2D(128, [3, 1], padding='same', activation='relu', name='conv4'),
    BatchNormalization(name='batch_norm4'),
    MaxPool2D([2, 1], padding='same', name='pool4'),
    Dropout(0.2, name='dropout4'),

    Conv2D(128, [4, 1], padding='valid', activation='relu', name='conv5.1'),
    Conv2D(96, [4, 1], padding='valid', activation='relu', name='conv5.2'),
    BatchNormalization(name='batch_norm5'),
    MaxPool2D([2, 1], padding='same', name='pool5'),

    Conv2D(64, [4, 1], padding='valid', activation='relu', name='conv6.1'),
    Conv2D(32, [4, 1], padding='valid', activation='relu', name='conv6.2'),
    BatchNormalization(name='batch_norm6'),
    MaxPool2D([2, 1], padding='valid', name='pool6'),  # <- changed padding to valid

    Conv2D(32, [4, 1], padding='valid', activation='relu', name='conv7.1'),
    Conv2D(32, [4, 1], padding='valid', activation='relu', name='conv7.2'),
    BatchNormalization(name='batch_norm7'),
    MaxPool2D([2, 1], padding='same', name='pool7'),
    Dropout(0.5, name='dropout7'),

    Flatten(name='flatten'),
    Dense(64, activation='relu', name='dense8'),
    Dense(1, activation='sigmoid', name='output'),
], name='CNN')

print('n layers:', len(model.layers))
model.summary()
```
