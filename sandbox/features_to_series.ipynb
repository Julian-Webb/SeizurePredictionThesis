```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {'Delta': [1, 6],
     'Theta': [2, 7],
     'Alpha': [3, 8],
     'Beta': [4, 9],
     'Gamma': [5, 10], }, index=['ch0', 'ch1']
)
df
```

```python
df.to_numpy().flatten()
```

```python
d = {}
for i, ch in enumerate(df.index):
    for band in df.columns:
        d[f'ch{i}_{band}'] = df.loc[ch, band]
d
```

```python
ser = pd.Series(d, name='value')
ser
```
